```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#read file It is a file with data added in blank cells
df = pd.read_excel("DS4PM F22 Homework_HP.xls")
```

```python
df = df.assign(count=df.isnull().sum(1))\
       .sort_values(['VIN#', 'count'])\
       .groupby('VIN#', as_index=0).head(1)\
       .drop('count', 1)
```

```python
df = df.drop(df[df['Mileage']<100].index)
```

```python
df
```

```python
df=df.dropna(subset=['Mileage'])
```

```python
df
```

```python
#Remove years before 2008
df = df.drop(df[df['Year']<2008].index)
```

```python
df
```

```python
df
```

```python
#Remove years after 2018
df = df.drop(df[df['Year']>2018].index)
df
```

```python
df
```

```python
#Standardize all Door data and add 4 in the blanks as it is majority
df['Doors'] = df['Doors'].replace(['Coupe'],'2')
df['Doors'] = df['Doors'].replace(['4 doors','4 Door', '4 Doors', '4 Door Sedan','5','four'],'4')
df
```

```python
df.to_excel('df_final.xls')
```

```python
#Made edits in excel for blank data and uploaded file (details in project file)
df = pd.read_excel("df_final.xls")
df
```

```python
#To understand the cars in data of car in a year and its mileage
#sns.relplot(x="Price", y="Year", hue="Mileage", data=df)
sns.set(rc = {'figure.figsize':(15,15)})
sns.scatterplot(data = df, x = "Price", y = "Year", hue = "Mileage")
```

```python
#As per previous scatterplot distribution I was able to understand the Good buying option based on cars which are new and median price with less mileage. 
#Car distribution per year
sns.set(style="ticks")
type_count = df['Year'].value_counts()
plt.figure(figsize=(14, 8))
ax = sns.barplot(x = type_count.index, y = type_count)
```

```python
#The above graph shows the amount of cars listed per year which shows the data has highest number of cars in 2017 and 2018
sns.set(style="ticks")
type_count = df['Mileage'].value_counts()
plt.figure(figsize=(14, 8))
ax = sns.barplot(x = type_count, y = type_count.index)
```
